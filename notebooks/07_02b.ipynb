```python
import difflib
```

```python
flines = 'Hello , How are you?,I am fine.'
glines = 'How are you Jerry?, I am doing Well.'
```

```python
d = difflib.Differ()
diff = d.compare(flines, glines)

for line in diff:
    print(line)
```
