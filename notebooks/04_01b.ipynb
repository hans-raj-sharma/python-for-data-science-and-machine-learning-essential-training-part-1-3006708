```python
import matplotlib.pyplot as plt
import seaborn as sns

from pandas import DataFrame
```

```python
data= {'Name':['Bob','Jessica','Mary','John','Mel','Tom','Jane','Jill','James','Jack'],
       'Age':[23, 20, 25, 22, 21, 25, 24, 27, 22, 24],
       'Gender':['M', 'F', 'F', 'M', 'M','M', 'F', 'F', 'M', 'M'],
       'Rank':[2, 1, 5, 4, 3, 6, 7, 8, 9, 10]}
df = DataFrame(data)
df
```

### Matplotlib's Bar Chart

```python
plt.bar(df['Name'], df['Age'])
plt.xlabel('Name')
plt.ylabel('Age')
plt.title('Age Comparision')
plt.show()
```

```python
plot = sns.barplot(x='Name', y='Age', data=df)
plot.set_title('Age Comparison')
plt.show()
```

### Line Plot Matplotlib

```python
plt.plot(df['Name'], df['Age'])
plt.xlabel('Names')
plt.ylabel('Age')
plt.title('Age Comparison')
plt.show()
```

### Line Plot Seaborn

```python
plot=sns.lineplot(data=df, x='Name', y='Age')
plot.set_title('Age Comparison')
plt.show()
```

### Pie Chart Matplotlib

```python
plt.pie(df['Age'], labels=df['Name'])
plt.title('Age Comparison')
plt.show()
```

### Pie Chart Seaborn

```python
colors = sns.color_palette('pastel')[0:6]
plt.pie(df['Age'], labels=df['Name'], colors=colors)
plt.title('Age Comparison')
plt.show()
```
