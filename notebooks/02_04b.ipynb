## Working with Missing Data in Pandas

```python
import numpy as np
import pandas as pd

from pandas import DataFrame
```

### Filling missing values using fillna(), replace() and interpolate()

```python
data ={'name':['John', 'Jane', 'Mary', 'Mark', 'Bob', 'Alice', 'Tom'],'age':[23, 25, 22, 24, 27, 21, 26],'gender':['M', 'F', 'F', 'M', 'M', 'F', 'M'],'rank':[1, 2, 3, 4, 5, 6, 7]} 

ranking_df = DataFrame(data)
ranking_df.iloc[2:5, 1]=np.nan
ranking_df.iloc[3:6, 3]=np.nan
ranking_df.iloc[3, :]=np.nan

ranking_df
```

```python
ranking_df.isnull()
```

```python
ranking_df.notnull()
```

```python
bool_series=pd.isnull(ranking_df['age'])
ranking_df[bool_series]
```

```python
ranking_df.fillna(0)
```

```python
ranking_df=ranking_df.ffill()
ranking_df
```

```python
ranking_df=ranking_df.bfill()
ranking_df
```

```python
ranking_df.dropna()
```

```python
ranking_df.dropna(how='all')
```

```python
ranking_df.dropna(axis=1)
```

```python
ranking_df.dropna(axis=0)
```
