### Defining plot color

```python
x=range(1,10)
y=[1,2,3,4,0.5,4,3,2,1]

plt.bar(x,y)
```

```python
wide=[0.5,0.5,0.5,0.9,0.9,0.9,0.5,0.5,0.5]
color=['red','red','red','blue','blue','blue','red','red','red']

plt.bar(x,y,width=wide,color=color)
```

### Customizing line styles

```python
x1=range(1,10)
y1=[9,8,7,6,5,4,3,2,1]
plt.plot(x,y)
plt.plot(x1,y1)
plt.show()
```

```python
plt.plot(x,y, ls='solid',lw=2)
plt.plot(x1,y1, ls='dashed',lw=2.5)
plt.show()
```

### Setting plot markers

```python
plt.plot(x,y,marker='1',mew=10)
plt.plot(x1,y1,marker='+',mew=11)
plt.show()
```
