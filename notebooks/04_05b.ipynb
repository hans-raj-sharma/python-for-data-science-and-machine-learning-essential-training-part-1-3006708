## Creating labels and annotations

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sb
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 12,4
sb.set_style('whitegrid')
```

### Labeling plot features
#### The functional method

```python
x=range(1,10)
y=[2,4,7,9,1,5,6,8,3]

plt.bar(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Bar Plot')
plt.show()
```

```python
z=[1,2,3,4,2.5]
veh_type=['car','truck','van','bus','suv']
plt.pie(z,labels=veh_type)
plt.show()
```

#### The object-oriented method

```python
address ='/workspaces/python-for-data-science-and-machine-learning-essential-training-part-1-3006708/data/mtcars.csv'

cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']

mpg = cars.mpg

fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

mpg.plot()

ax.set_xticks(range(32))

ax.set_xticklabels(cars.car_names,rotation=90, fontsize='medium')
ax.set_title('Miles Per Gallon of Cars in mtcars')

ax.set_xlabel('Car Names')
ax.set_ylabel('Miles Per Gallon')
plt.show()
```

### Adding a legend to your plot
#### The functional method

```python
plt.pie(z)
plt.legend(veh_type,loc='best')
plt.show()
```

#### The object-oriented method

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

mpg.plot()

ax.set_xticks(range(32))

ax.set_xticklabels(cars.car_names,rotation=90, fontsize='medium')
ax.set_title('Miles Per Gallon of Cars in mtcars')

ax.set_xlabel('Car Names')
ax.set_ylabel('Miles Per Gallon')

ax.legend(loc='best')
```

### Annotating your plot

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

mpg.plot()

ax.set_xticks(range(32))

ax.set_xticklabels(cars.car_names,rotation=90, fontsize='medium')
ax.set_title('Miles Per Gallon of Cars in mtcars')

ax.set_xlabel('Car Names')
ax.set_ylabel('Miles Per Gallon')

ax.set_ylim(0,40)

ax.annotate('Toyota Corolla', xy=(19,33.9), xytext=(21,35), arrowprops=dict(facecolor='black', shrink=0.05))

ax.legend(loc='best')
```
