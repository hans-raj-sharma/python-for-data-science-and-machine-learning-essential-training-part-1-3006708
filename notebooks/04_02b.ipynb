```python
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams
```

### Creating a line chart from a list object

```python
x=range(1,10)
y=[1,4,9,16,25,36,49,64,81]
plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('X vs Y')
plt.show()
```
