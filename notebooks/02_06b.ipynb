```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
```

### Concatenating data

```python
DF_obj=DataFrame(np.arange(49).reshape(7,7))
DF_obj
```

```python
DF_obj2=DataFrame(np.arange(16).reshape(4,4))
DF_obj2
```

```python
pd.concat([DF_obj,DF_obj2],axis=1)
```

```python
pd.concat([DF_obj,DF_obj2])
```

### Transforming data
#### Dropping data

```python
DF_obj.drop([1,2])
```

```python
DF_obj.drop([1,2,3],axis=1)
```

#### Adding data

```python
series_obj=Series(np.arange(10))
series_obj.name='my_series'
series_obj
```

```python
variable_add = DataFrame.join(DF_obj, series_obj)
variable_add
```

```python
added_datatable = pd.concat([variable_add,variable_add], ignore_index=False)
added_datatable
```

```python
added_datatable = pd.concat([variable_add,variable_add], ignore_index=True)
added_datatable
```

#### Sorting data

```python
DF_sorted = DF_obj.sort_values(by=[5], ascending=False)
DF_sorted
```
